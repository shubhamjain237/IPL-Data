### Top 5 Run getters for Every Season

### Points Table

```python
year = np.sort(df_year)
df_points_table = pd.DataFrame([['','','','','']])
df_points_dict = dict()
for year in year:
    df_points_dict[year] = dict()
    df_partial = df_matches[df_matches['season']==year]
    df_partial.reset_index(inplace=True,drop=True)
    teams = df_partial['team1'].unique()
    if year==2011:
        total_matches_per_team = 14
        df_partial = df_partial.loc[:69,]
    else:
        total_matches_per_team = 2*(len(teams)-1)
        df_partial = df_partial.loc[:len(teams)*(len(teams)-1)-1,]
    for team in teams:
        df_points_dict[year][team] = [0,0,0] 
    for i in range(len(df_partial)):
        if df_partial.loc[i,'winner']=='None':
            df_points_dict[year][df_partial.loc[i,'team1']][1] = df_points_dict[year][df_partial.loc[i,'team1']][1]+1
            df_points_dict[year][df_partial.loc[i,'team2']][1] = df_points_dict[year][df_partial.loc[i,'team2']][1]+1
        else:
            df_points_dict[year][df_partial.loc[i,'winner']][0] = df_points_dict[year][df_partial.loc[i,'winner']][0]+1
    for team in teams:
        df_points_dict[year][team][2] = total_matches_per_team-df_points_dict[year][team][0]-df_points_dict[year][team][1]
    for team in teams:
        df_points_table = df_points_table.append([[year,team,df_points_dict[year][team][0],df_points_dict[year][team][2],
                                                  df_points_dict[year][team][1]]],ignore_index=True)
df_points_table.columns = ['Year','Team','Wins','Loss','N/R']
df_points_table.drop(index=0,inplace=True)
df_points_table.astype({'Year':int,'Wins':int,'Loss':int,'N/R':int})
df_points_table.sort_values(by=['Year','Wins','N/R'],ascending=[True,False,False],inplace=True)
df_points_table.reset_index(drop=True,inplace=True)
df_points_table['Points'] = 2*df_points_table['Wins']+1*df_points_table['N/R']
df_points_table.to_csv(r'C:/Users/lenovo/Desktop/IPL/points_tabls.csv',index=False)
```

### Highest Score in every Season

```python
year = np.sort(df_year)
df_highest_runs = pd.DataFrame([['','','','']])
for year in year:
    batsman_list = []
    score_list = []
    match_id_list = []
    df_partial = df_full[df_full['Season']==year][['Season','batsman','batsman_runs','match_id']]
    df_batsman_group = df_partial.groupby(['batsman','match_id'])
    for batsman,data in df_batsman_group:
        batsman_list.append(batsman[0])
        score_list.append(data['batsman_runs'].sum())
        match_id_list.append(batsman[1])
    zipped_object = zip(batsman_list,score_list,match_id_list)
    zipped_list = sorted(zipped_object,key=lambda z:z[1],reverse=True)
    zipped_list = zipped_list[:5]
    counter=1
    for i,j,k in zipped_list:
        df_highest_runs = df_highest_runs.append([[year,str(counter)+' '+i,j,k]],ignore_index=True)
        counter = counter+1
df_highest_runs.drop(index=0,inplace=True)
df_highest_runs.reset_index(inplace=True,drop=True)
df_highest_runs.columns = ['Year','Player','Highest Score','Match Id']
df_highest_runs.to_csv(r'C:/Users/lenovo/Desktop/IPL/highest_runs.csv',index=False)
```

### Hat Tricks

```python
year = np.sort(df_year)
df_hat_trick = pd.DataFrame([['','','']])
bowler_hat = dict()
for year in year:
    bowler_hat[year] = dict()
    df_partial = df_full[(df_full['Season']==year) & ((df_full['wide_runs']==0)
                                                      | ((df_full['wide_runs']!=0) & (df_full['dismissal_kind']=='stumped')))]
    df_group = df_partial.groupby(['bowler','match_id'])
    for i,j in df_group:
        wickets_list = []
        for k in j['dismissal_kind']:
            wickets_list.append(k)
        hatrick_found = False
        counter = 0
        count_wickets = 0
        for u in range(len(wickets_list)-2):
            count_wickets = 0
            for v in range(3):
                if pd.isnull(wickets_list[u+v]):
                    hatrick_found = False
                    break
                elif wickets_list[u+v] in ['caught','bowled','lbw','caught and bowled','stumped','hit wicket']:
                    if not hatrick_found:
                        count_wickets = count_wickets+1
                        if count_wickets==3:
                            count_wickets = 0
                            hatrick_found = True
                            counter = counter+1
                    else:
                        break
        if counter!=0:
            if i[0] not in bowler_hat[year]:
                bowler_hat[year][i[0]] = 1
            else:
                bowler_hat[year][i[0]] = bowler_hat[year][i[0]]+1
    for i in bowler_hat[year]:
        df_hat_trick = df_hat_trick.append([[year,i,bowler_hat[year][i]]],ignore_index=True)
df_hat_trick.drop(index=0,inplace=True)
df_hat_trick.reset_index(drop=True,inplace=True)
df_hat_trick.columns = ['Year','Player','Number of Hat Tricks']
df_hat_trick.to_csv(r'C:/Users/lenovo/Desktop/IPL/hattricks.csv',index=False)
```

### Best Bowling Average per season

```python
df_best_bowling_average = pd.DataFrame([['','','','','']])
year = np.sort(df_year)
for year in year:
    runs_conceeded = []
    wickets_taken = []
    bowler_name = []
    average_list = []
    df_partial = df_full[df_full['Season']==year]
    df_grouped = df_partial.groupby('bowler')
    for index,entry in df_grouped:
        entry = entry[(entry['bye_runs']==0) & (entry['legbye_runs']==0)]
        if len(entry)>=12:
            bowler_name.append(index)
            runs_conceeded.append(entry['total_runs'].sum())
            wickets_taken.append(len(entry[(entry['dismissal_kind']=='caught') | (entry['dismissal_kind']=='bowled') | 
                                (entry['dismissal_kind']=='lbw') | (entry['dismissal_kind']=='caught and bowled') | 
                                (entry['dismissal_kind']=='stumped') | (entry['dismissal_kind']=='hit wicket')]))
            average_list.append(runs_conceeded[-1]/wickets_taken[-1])
    zipped_object = zip(bowler_name,runs_conceeded,wickets_taken,average_list)
    zipped_object = sorted(zipped_object,key=lambda z:z[3])
    zipped_object = zipped_object[:5]
    for i,j,k,l in zipped_object:
        df_best_bowling_average = df_best_bowling_average.append([[year,i,j,k,l]],ignore_index=True)
df_best_bowling_average.drop(index=0,inplace=True)
df_best_bowling_average.reset_index(inplace=True,drop=True)
df_best_bowling_average.columns = ['Year','Player','Runs Conceeded','Wickets','Average']
df_best_bowling_average.to_csv(r'C:/Users/lenovo/Desktop/IPL/best_bowling_average.csv',index=False)
```
